```
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
from datasets import Dataset
import os
import wandb
import torch.nn.functional as F
from transformers import TrainerCallback

class MetricsCallback(TrainerCallback):
    def __init__(self, trainer, test_dataset, label_mapping, metrics_file="training_scores.json"):
        self.trainer = trainer
        self.test_dataset = test_dataset
        self.label_mapping = label_mapping
        self.metrics_file = metrics_file

    def on_epoch_end(self, args, state, control, **kwargs):
        # Evaluate the model at the end of each epoch
        epoch_metrics = evaluate_model(self.trainer, self.test_dataset, self.label_mapping, self.metrics_file)
        print(f"Epoch {state.epoch} Metrics:", epoch_metrics)
        plot_metrics()
        wandb.log(epoch_metrics)

# Load dataset
def load_data(filepath="/content/news-article-categories.csv"):
    with open("/content/news-article-categories.csv", "r", encoding="utf-8") as f:
      for _ in range(5):
        print(f.readline())  # Prints first 5 lines

    df = pd.read_csv("/content/news-article-categories.csv",
                 encoding="utf-8",         # Handle special characters
                 on_bad_lines="skip",      # Skip problematic rows
                 quoting=3)
    df = df.rename(columns={'category': 'label_name', 'title': 'title', 'body': 'description'})
    label_mapping = {name: idx for idx, name in enumerate(df['label_name'].unique())}
    df['label_id'] = df['label_name'].map(label_mapping)
    return df, label_mapping

# Prepare dataset for training
def prepare_dataset(df, tokenizer, max_length=512):
    df['text'] = df['title'].astype(str) + " " + df['description'].astype(str)
    df['text'] = df['text'].fillna("")
    dataset = Dataset.from_pandas(df[['text', 'label_id']])

    def tokenize_function(examples):
        encoding = tokenizer(
            examples["text"],
            padding='max_length',
            truncation=True,
            max_length=max_length
        )
        encoding["labels"] = examples["label_id"]
        return encoding

    return dataset.map(tokenize_function, batched=True)

# Train model
def train_model(train_dataset, eval_dataset, model, tokenizer, label_mapping, output_dir="./results"):
    training_args = TrainingArguments(
        output_dir=output_dir,
        num_train_epochs=10,  # Number of epochs
        per_device_train_batch_size=32,
        per_device_eval_batch_size=32,
        evaluation_strategy="epoch",  # Ensure evaluation happens after every epoch
        save_strategy="epoch",
        logging_dir='./logs',
        load_best_model_at_end=True
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=eval_dataset
    )

    # Add custom callback for per-epoch evaluation
    trainer.add_callback(MetricsCallback(trainer, eval_dataset, label_mapping))

    trainer.train()
    trainer.save_model(output_dir)
    tokenizer.save_pretrained(output_dir)
    return trainer

# Evaluate model and save metrics
def evaluate_model(trainer, test_dataset, label_mapping, metrics_file="training_scores.json"):
    predictions = trainer.predict(test_dataset)
    preds_tensor = torch.tensor(predictions.predictions, dtype=torch.float)
    preds = np.argmax(predictions.predictions, axis=-1)
    test_labels = test_dataset['label_id']

    report = classification_report(test_labels, preds, output_dict=True)
    confusion_mat = confusion_matrix(test_labels, preds)

# Ensure labels are integer indices
    labels_tensor = torch.tensor(test_labels, dtype=torch.long)
    loss_value = F.cross_entropy(preds_tensor, F.one_hot(labels_tensor, num_classes=preds_tensor.shape[1]).float()).item()

    current_metrics = {
        "accuracy": accuracy_score(test_labels, preds),
        "precision": precision_score(test_labels, preds, average='weighted'),
        "recall": recall_score(test_labels, preds, average='weighted'),
        "f1_score": f1_score(test_labels, preds, average='weighted'),
        "loss": loss_value,
        "confusion_matrix": confusion_mat.tolist()
    }

    # Save metrics
    if os.path.exists(metrics_file):
        with open(metrics_file, "r") as f:
            try:
                all_metrics = json.load(f)
            except json.JSONDecodeError:
                all_metrics = []
    else:
        all_metrics = []

    all_metrics.append(current_metrics)
    with open(metrics_file, "w") as f:
        json.dump(all_metrics, f, indent=4)

    return current_metrics

# Plot training performance
def plot_metrics(metrics_file="training_scores.json"):
    if not os.path.exists(metrics_file):
        print("No training data found.")
        return

    with open(metrics_file, "r") as f:
        try:
            data = json.load(f)
        except json.JSONDecodeError:
            print("Error reading metrics file.")
            return

    if not data:
        print("No valid training data found.")
        return

    # Extract values across iterations
    epochs = list(range(1, len(data) + 1))
    acc = [m["accuracy"] for m in data]
    loss = [m.get("loss", None) for m in data]  # Use None if 'loss' is missing
    prec = [m["precision"] for m in data]
    rec = [m["recall"] for m in data]
    f1 = [m["f1_score"] for m in data]

    # Compute Mean & Std Dev
    summary_metrics = {
        "accuracy_mean": np.mean(acc),
        "accuracy_std": np.std(acc),
        "loss_mean": np.mean(loss),
        "loss_std": np.std(loss),
        "precision_mean": np.mean(prec),
        "precision_std": np.std(prec),
        "recall_mean": np.mean(rec),
        "recall_std": np.std(rec),
        "f1_score_mean": np.mean(f1),
        "f1_score_std": np.std(f1),
    }

    print("\n===== Summary Metrics (Averages Over Training Runs) =====")
    print(f"Accuracy:  Mean = {summary_metrics['accuracy_mean']:.4f},  Std = {summary_metrics['accuracy_std']:.4f}")
    print(f"Loss:      Mean = {summary_metrics['loss_mean']:.4f},  Std = {summary_metrics['loss_std']:.4f}")
    print(f"Precision: Mean = {summary_metrics['precision_mean']:.4f},  Std = {summary_metrics['precision_std']:.4f}")
    print(f"Recall:    Mean = {summary_metrics['recall_mean']:.4f},  Std = {summary_metrics['recall_std']:.4f}")
    print(f"F1 Score:  Mean = {summary_metrics['f1_score_mean']:.4f},  Std = {summary_metrics['f1_score_std']:.4f}")
    print("========================================================\n")
    wandb.log(summary_metrics)
    # 📌 Plot Accuracy vs Epochs
    plt.figure(figsize=(10, 5))
    plt.plot(epochs, acc, label='Accuracy', color='blue', marker='o')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title('Model Accuracy Over Epochs')
    plt.legend()
    plt.grid()
    wandb.log({"Accuracy Plot": wandb.Image(plt)})
    plt.show()


    # 📌 Plot Loss vs Epochs
    plt.figure(figsize=(10, 5))
    plt.plot(epochs, loss, label='Loss', color='red', marker='o')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Model Loss Over Epochs')
    plt.legend()
    plt.grid()
    wandb.log({"Loss Plot": wandb.Image(plt)})
    plt.show()

    # 📌 Plot Precision, Recall, and F1-score vs Epochs
    plt.figure(figsize=(10, 5))
    plt.plot(epochs, prec, label='Precision', color='green', marker='o')
    plt.plot(epochs, rec, label='Recall', color='orange', marker='o')
    plt.plot(epochs, f1, label='F1-score', color='purple', marker='o')
    plt.xlabel('Epochs')
    plt.ylabel('Score')
    plt.title('Precision, Recall, and F1-score Over Epochs')
    plt.legend()
    plt.grid()
    wandb.log({"Precision-Recall-F1 Plot": wandb.Image(plt)})
    plt.show()

    # 📌 Plot Confusion Matrix (last epoch)
    last_conf_matrix = np.array(data[-1]["confusion_matrix"])
    plt.figure(figsize=(10, 7))
    sns.heatmap(last_conf_matrix, annot=True, fmt="d", cmap="Blues")
    plt.xlabel("Predicted Labels")
    plt.ylabel("True Labels")
    plt.title("Confusion Matrix (Final Epoch)")
    wandb.log({"Confusion Matrix": wandb.Image(plt)})
    plt.show()

# Classify new articles
def classify_article(title, description, model_path="./final_model", label_mapping=None):
    tokenizer = AutoTokenizer.from_pretrained(model_path)
    model = AutoModelForSequenceClassification.from_pretrained(model_path)
    text = title + " " + description
    inputs = tokenizer([text], padding='max_length', truncation=True, max_length=512, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)
        logits = outputs.logits
        predicted_class = torch.argmax(logits, dim=1).item()
    if label_mapping is None:
        raise ValueError("label_mapping is required for category name conversion.")
    id_to_label = {idx: name for name, idx in label_mapping.items()}
    category_label = id_to_label.get(predicted_class, f"Unknown ({predicted_class})")
    print("Predicted Category:", category_label)
    return category_label

# Main pipeline
def main():
    model_name = "distilbert-base-uncased"

    # Load dataset (FULL DATA, NO RANDOM SAMPLING)
    df, label_mapping = load_data("news-article-categories.csv")

    # Train-test split
    train_df, temp_df = train_test_split(df, test_size=0.3, random_state=42)
    eval_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42)

    # Load tokenizer & model
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=len(label_mapping))

    # Prepare datasets
    train_dataset = prepare_dataset(train_df, tokenizer)
    eval_dataset = prepare_dataset(eval_df, tokenizer)
    test_dataset = prepare_dataset(test_df, tokenizer)

    # Train model
    trainer = train_model(train_dataset, eval_dataset, model, tokenizer, label_mapping)

    # Evaluate model
    metrics = evaluate_model(trainer, test_dataset, label_mapping)
    print("Final Metrics:", metrics)

    # Plot performance
    plot_metrics()

    # Save final model
    trainer.save_model("./final_model")
    tokenizer.save_pretrained("./final_model")

    print("Training complete. Model saved.")

if __name__ == "__main__":
    main()
```
